````python
# more bivariate analysis
db.head(5)
````

````python

# Correlation heatmap
numeric_cols = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
sns.heatmap(db[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Numeric Features')
plt.show()
````

````python
binary_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

for col in binary_cols:
    sns.boxplot(data=db, x=col, y='price')
    plt.title(f'Price by {col.capitalize()}')
    plt.show()
````

````python
#pairplot for numerical features
custom_palette = {
    'price': 'blue',
    'area': 'green',
    'bedrooms': 'orange',
    'bathrooms': 'purple',
    'stories': 'red',
    'parking': 'brown'
}

sns.pairplot(db[numeric_cols], diag_kind='kde', markers='o',palette=custom_palette)
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.xlabel('Numerical Features')
plt.ylabel('Numerical Features')
plt.legend(title='Numerical Features', loc='upper right')
plt.tight_layout()
plt.subplots_adjust(top=0.9)  # Adjust the top to make room for the title
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.grid(True)
plt.show()
````

````python
# Heatmap of counts
ct = pd.crosstab(db['basement'], db['hotwaterheating'])
sns.heatmap(ct, annot=True, cmap='Blues')
plt.title('Basement vs Hot Water Heating')
plt.show()
````

Get creative and enhance your dataset! 🚀

````python
#add your code here
#creating new features 
db['price_per_area'] = db['price'] / db['area']
db['bed_rooms_bathrooms'] = db['bedrooms'] + db['bathrooms']
````

````python
#ordinal encoding for 'furnishingstatus'
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
db['furnishingstatus'] = ordinal_encoder.fit_transform(db[['furnishingstatus']])
# label encoding for categorical features
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for col in categorical_cols:
    db[col] = label_encoder.fit_transform(db[col])
# Display the first few rows of the modified DataFrame
db.head(10)
````

````python
#change the data type of stories to float
db['stories'] = db['stories'].astype(float)
````

````python
#scaling numerical features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_cols = ['stories', 'parking', 'price_per_area', 'bed_rooms_bathrooms']
db[numerical_cols] = scaler.fit_transform(db[numerical_cols])
````
